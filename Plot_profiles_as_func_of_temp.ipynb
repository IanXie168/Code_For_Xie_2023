```python
node_qc_profile_list = [];  node_qr_profile_list = [];
node_qi_profile_list = [];  node_qg_profile_list = [];
node_qs_profile_list = []


node_qcn_profile_list = [];  node_qrn_profile_list = [];
node_qin_profile_list = [];  node_qwfa_profile_list = [];
node_qifa_profile_list = []; node_ratio_profile_list = []

for keys, index in dic_cluster.items():
    
    node_qc_profile = np.nanmean(qc_profile[index], axis=0) * 1000
    node_qr_profile = np.nanmean(qr_profile[index], axis=0) * 1000
    node_qs_profile = np.nanmean(qs_profile[index], axis=0) * 1000
    node_qi_profile = np.nanmean(qi_profile[index], axis=0) * 1000
    node_qg_profile = np.nanmean(qg_profile[index], axis=0) * 1000
    node_ratio_profile = np.nanmean(ratio_profile[index], axis=0)
    
    node_qc_profile_list.append(node_qc_profile); node_qr_profile_list.append(node_qr_profile);
    node_qi_profile_list.append(node_qi_profile); node_qg_profile_list.append(node_qg_profile);
    node_qs_profile_list.append(node_qs_profile); node_ratio_profile_list.append(node_ratio_profile)

    node_qcn_profile  = np.nanmean(qcn_profile[index], axis=0)
    node_qrn_profile  = np.nanmean(qrn_profile[index], axis=0)
    node_qin_profile  = np.nanmean(qin_profile[index], axis=0)
    node_qwfa_profile = np.nanmean(qwfa_profile[index], axis=0)
    node_qifa_profile = np.nanmean(qifa_profile[index], axis=0)
    
    node_qcn_profile_list.append(node_qcn_profile); node_qrn_profile_list.append(node_qrn_profile);
    node_qin_profile_list.append(node_qin_profile); node_qwfa_profile_list.append(node_qwfa_profile);
    node_qifa_profile_list.append(node_qifa_profile);
```
